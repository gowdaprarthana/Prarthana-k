````python
import sqlite3
import os
from langchain.agents import initialize_agent, Tool
from langchain_core.tools import tool
from langchain_groq import ChatGroq
from google.colab import files

# ========== Set up Groq API ==========
os.environ["GROQ_API_KEY"] = "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"
llm = ChatGroq(
    groq_api_key=os.environ["GROQ_API_KEY"],
    model_name="llama3-70b-8192"
)

# ========== Upload or Create SQLite DB ==========
print("üì§ Upload your SQLite (.db) file or press 'Skip' to use the demo database.")
uploaded_files = files.upload()
if uploaded_files:
    db_path = list(uploaded_files.keys())[0]
    conn = sqlite3.connect(db_path)
    print(f"‚úÖ Connected to uploaded DB: {db_path}")
else:
    conn = sqlite3.connect("demo.db")
    cursor = conn.cursor()
    cursor.executescript("""
    DROP TABLE IF EXISTS customers;
    DROP TABLE IF EXISTS orders;

    CREATE TABLE customers (
        customer_id INTEGER PRIMARY KEY,
        name TEXT,
        city TEXT
    );

    CREATE TABLE orders (
        order_id INTEGER PRIMARY KEY,
        customer_id INTEGER,
        order_date TEXT,
        FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
    );

    INSERT INTO customers (name, city) VALUES ('Alice', 'Chennai'), ('Bob', 'Mumbai');
    INSERT INTO orders (customer_id, order_date) VALUES (1, '2023-05-01'), (2, '2023-06-01');
    """)
    conn.commit()
    print("‚úÖ Using demo DB with sample customer/order data.")

cursor = conn.cursor()

# ========== Tool: EXPLAIN QUERY PLAN ==========
@tool
def explain_sql(query: str) -> str:
    """Run EXPLAIN QUERY PLAN on SQL using SQLite and return the plan."""
    try:
        cursor.execute("EXPLAIN QUERY PLAN " + query)
        plan = cursor.fetchall()
        return "üìò Execution plan: " + "; ".join([str(p[3]) for p in plan[:4]])
    except Exception as e:
        return f"‚ùå Error in EXPLAIN: {str(e)}"

# ========== Tool: Optimize SQL using LLM ==========
@tool
def optimize_sql(query: str) -> str:
    """Use LLM to analyze and optimize the SQL query."""
    prompt = (
        f"Optimize the following SQL query:\n\n"
        f"```sql\n{query}\n```\n\n"
        "Return only:\n"
        "1. Up to 4 lines of explanation.\n"
        "2. The optimized SQL query in a code block.\n"
    )
    return llm.invoke(prompt).content

# ========== Agent Initialization ==========
tools = [explain_sql, optimize_sql]
agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent_type="zero-shot-react-description",
    verbose=False
)

# ========== Interaction Loop ==========
print("\nüß† SQL Optimizer Agent Ready!\nType your SQL query or type 'exit' to quit.")
while True:
    user_query = input("\nüîç Enter your SQL query (or type 'exit' to stop):\n")
    if user_query.strip().lower() == "exit":
        print("üëã Exiting. Have a great day!")
        break
    try:
        print("\nüß† Thinking...\n")
        result = agent.run(f"Explain and optimize this SQL query: {user_query}")
        print(result)
    except Exception as e:
        print("‚ùå Error:", e)
````
